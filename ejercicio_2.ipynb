### Imports

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics.pairwise import cosine_similarity
```

### 1. Compara el rendimiento de la SVM con kernel coseno, lineal y RBF.

```python
newsgroups_train = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
newsgroups_test = fetch_20newsgroups(subset='test', remove=('headers', 'footers', 'quotes'))
# newsgroups_train = fetch_20newsgroups(subset='train')
# newsgroups_test = fetch_20newsgroups(subset='test')

vectorizer = TfidfVectorizer()
vectors_train = vectorizer.fit_transform(newsgroups_train.data)
vectors_test = vectorizer.transform(newsgroups_test.data)
```

1a) Kernel coseno

```python
# Compute the cosine similarity matrix
cosine_kernel_train = cosine_similarity(vectors_train)
cosine_kernel_test = cosine_similarity(vectors_test, vectors_train)

# define SVM classifier with cosine kernel
classifier = SVC(kernel='precomputed')
classifier.fit(cosine_kernel_train, newsgroups_train.target)

# predict test data
pred = classifier.predict(cosine_kernel_test)
print("NOTE: The performance is quite bad, because the header, footer and quotes are removed \
      (as recommended at https://scikit-learn.org/0.19/datasets/twenty_newsgroups.html for a more realistic training).")
print("Accuracy: ", metrics.accuracy_score(newsgroups_test.target, pred))
print("Precision: ", metrics.precision_score(newsgroups_test.target, pred, average='macro'))
print("Recall: ", metrics.recall_score(newsgroups_test.target, pred, average='macro'))
print("F1: ", metrics.f1_score(newsgroups_test.target, pred, average='macro'))
```

1b) Kernel lineal

```python
# define SVM classifier with cosine kernel
classifier = SVC(kernel='linear')
classifier.fit(vectors_train, newsgroups_train.target)

# predict test data
pred = classifier.predict(vectors_test)
print("NOTE: The performance is quite bad, because the header, footer and quotes are removed \
      (as recommended at https://scikit-learn.org/0.19/datasets/twenty_newsgroups.html for a more realistic training).")
print("Accuracy: ", metrics.accuracy_score(newsgroups_test.target, pred))
print("Precision: ", metrics.precision_score(newsgroups_test.target, pred, average='macro'))
print("Recall: ", metrics.recall_score(newsgroups_test.target, pred, average='macro'))
print("F1: ", metrics.f1_score(newsgroups_test.target, pred, average='macro'))
```

1c) Kernel RBF

```python
# define SVM classifier with cosine kernel
classifier = SVC(kernel='rbf')
classifier.fit(vectors_train, newsgroups_train.target)

# predict test data
pred = classifier.predict(vectors_test)
print("NOTE: The performance is quite bad, because the header, footer and quotes are removed \
      (as recommended at https://scikit-learn.org/0.19/datasets/twenty_newsgroups.html for a more realistic training).")
print("Accuracy: ", metrics.accuracy_score(newsgroups_test.target, pred))
print("Precision: ", metrics.precision_score(newsgroups_test.target, pred, average='macro'))
print("Recall: ", metrics.recall_score(newsgroups_test.target, pred, average='macro'))
print("F1: ", metrics.f1_score(newsgroups_test.target, pred, average='macro'))
```
